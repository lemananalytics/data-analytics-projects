```python
import pandas as pd
import numpy as np
data = {
    'Tr_ID': [101, 102, 103, 103, 105, 106, 107, 108, 109, 110],
    'Tarix': ['2026-05-01', '2026-05-02', '2026/05/03', '2026-05-03', '2026-05-05', '2026-05-06', None, '2026-05-08', '2026-05-09', '2026-05-10'],
    'Mehsul_Kateqoriyasi': ['Elektronika', ' geyim ', 'Elektronika', 'Elektronika', 'Məişət', 'Geyim', 'Elektronika', 'Məişət', None, 'Geyim'],
    'Satilan_Say': [2, -5, 1, 1, 3, None, 5, 2, 4, 1],
    'Vahid_Qiymet': [150, 40, 200, 200, 80, 50, 120, None, 30, 60],
    'Region': ['Bakı', 'Gəncə', 'Bakı', 'Bakı', 'Sumqayıt', 'Bakı', 'Gəncə', 'Sumqayıt', 'Bakı', 'Bakı']
}
df = pd.DataFrame(data)
df
```

```python
print("---Təmizlənmiş data---")
df = df.drop_duplicates(subset=['Tr_ID'])
df = df.dropna(subset=['Tarix', 'Mehsul_Kateqoriyasi','Tr_ID'])

df['Tarix'] = pd.to_datetime(df['Tarix'], format='mixed')
df['Mehsul_Kateqoriyasi']=df['Mehsul_Kateqoriyasi'].str.strip().str.capitalize()
df['Satilan_Say']=df['Satilan_Say'].clip(1,100).fillna(1).astype(int)
vahid_qiymet_ortalama=np.nanmedian(df['Vahid_Qiymet'])
df['Vahid_Qiymet']=df['Vahid_Qiymet'].fillna(vahid_qiymet_ortalama)
df
```

```python
print("---Filtrləmə---")
a=df[(df['Region']=='Bakı') & (df['Vahid_Qiymet']>100)]
print("Bakı regionunda vahid qiyməti 100-dən çox olan məhsullar:")
a
```

```python
print("---Yeni sütun yaratma---")
df['Gelir']=df['Satilan_Say']*df['Vahid_Qiymet']
print("Gəlir sütunu əlavə edilmiş data:")
df
```

```python
print("---Qruplaşdırma---")
b=df.groupby('Mehsul_Kateqoriyasi').agg({'Satilan_Say': 'sum', 'Gelir': 'sum'})
print(b)
```
